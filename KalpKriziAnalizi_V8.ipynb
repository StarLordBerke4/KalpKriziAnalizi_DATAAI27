##  Kalp Krizi Hakknda

* Kalp krizi, **miyokard enfarkts** olarak da adlandrlr, kalp kasnn bir ksmna yeterli kan gitmediinde meydana gelir.
* Kan akn geri kazandrmak iin tedavi edilmeden geen sre ne kadar uzun olursa, kalp kasna verilen hasar da o kadar byk olur.
* **Koroner arter hastal (KAH)**, kalp krizinin balca nedenidir.
* KAH, kalbi besleyen atardamarlarn (koroner arterler) genellikle ya ve kolesterol birikimiyle daralmas veya tkanmas sonucu oluur. Bu duruma **ateroskleroz** denir.
* Belirtiler genellikle gste ar veya rahatszlk, nefes darl, kola (zellikle sol kola), eneye, srta veya mideye yaylan ar eklinde kendini gsterir. Baz durumlarda ise belirtiler daha hafif olabilir veya hi grlmeyebilir (sessiz kalp krizi).
* Kalp krizi riskini artran faktrler arasnda **yksek tansiyon, yksek kolesterol, diyabet, obezite, sigara kullanm, fiziksel hareketsizlik ve stres** yer alr. Genetik yatknlk da nemli bir risk faktrdr.
* Kalp krizi geiren kiilerde hzl tbbi mdahale hayati nem tar. Acil tedavi genellikle kan phtsn zc ilalar (trombolitikler), balon anjiyoplasti veya stent yerletirme gibi yntemleri ierir.
* Kalp krizinden korunmak iin **salkl bir yaam tarz benimsemek, dzenli egzersiz yapmak, dengeli beslenmek ve risk faktrlerini ynetmek** byk nem tar.

##  Standardizasyon

* Veri bilimi standardizasyonu, verilerin tutarlln ve kalitesini artrmak amacyla kullanlr. Bu sre, farkl kaynaklardan gelen verilerin uyumlu hale getirilmesini salar, bylece analiz sreleri daha verimli ve gvenilir olur.

```python
# StandardScaler snf, verileri standart lee dntrmek iin kullanlr
# Yani her bir zellii (stunu), ortalamas 0, standart sapmas 1 olacak ekilde dntrr
scaler = StandardScaler()

# Oluturulan scaler nesnesi arldnda, ierii (rnein parametre ayarlar) grntlenir
scaler
```

```python
# StandardScaler nesnesi kullanlarak saysal veriler standardize ediliyor (lekleniyor)
# df[numeric_list[:-1]] ifadesi: 'output' stunu hari, listedeki tm saysal stunlar seer
# numeric_list[:-1] -> listenin son eleman hari (output hari) olan ksm
# scaler.fit_transform(...) -> veriyi hem renir (fit), hem de dntrr (transform)
scaled_array = scaler.fit_transform(df[numeric_list[:-1]])
```

```python
# scaled_array deikeni, StandardScaler ile leklendirilmi saysal verileri ierir
# Bu veri, her stunun ortalamas 0 ve standart sapmas 1 olacak ekilde dntrlmtr
# Bu satr altrldnda, scaled_array'in ierii (NumPy dizisi olarak) ekrana yazdrlr
scaled_array
```

##  Kutu Grafii Analizi

* Kutu grafii, veri bilimi alannda verilerin dalmn ve zet istatistiklerini grselletirmek iin kullanlan etkili bir aratr. zellikle, bir veri kmesinin merkezi eilimi (ortalama veya medyan), yaylm (eyrekler aras aralk) ve olas aykr deerleri hzl bir ekilde anlamaya yardmc olur. Kutu grafikleri, farkl gruplar arasndaki karlatrmalar kolaylatrarak, verilerdeki farkllklar ve benzerlikleri belirlemeye olanak tanr. Bu sayede, karar verme srelerinde daha bilinli ve veri odakl yaklamlar gelitirilir.

```python
# scaled_array verisi (NumPy dizisi), stun adlar belirtilerek pandas DataFrame'e dntrlyor
# columns = numeric_list[:-1] -> 'output' hari tm saysal stun adlar ekleniyor
df_dummy = pd.DataFrame(scaled_array, columns=numeric_list[:-1])

# Oluturulan df_dummy veri erevesinin ilk 5 satr ekranda gsteriliyor
df_dummy.head()
```

```python
# Daha nce oluturulan df_dummy (leklenmi saysal veriler) ile,
# orijinal veri setindeki 'output' stunu birletiriliyor
# axis=1 ifadesi: Birletirme ilemi stun baznda (yan yana) yaplacak demektir
df_dummy = pd.concat([df_dummy, df.loc[:, "output"]], axis=1)

# Birletirilen yeni DataFrame'in ilk 5 satr grntleniyor
df_dummy.head()
```

```python
# df_dummy veri seti, pd.melt() fonksiyonu ile 'long format'a dntrlyor
# id_vars="output"  'output' stunu sabit tutulur, dier stunlar "features" ad altnda eritilir
# var_name="features"  Stun adlar bu yeni stunda toplanr
# value_name="value"  Her stundaki deerler bu yeni stunda yer alr
data_melted = pd.melt(df_dummy, 
                      id_vars="output",        # 'output' sabit kalacak (snf etiketi)
                      var_name="features",     # Dier stun adlar bu stunda toplanacak
                      value_name="value")      # zelliklerin deerleri bu stunda olacak

# Elde edilen long-form veri setinin ilk 20 satr grntleniyor
data_melted.head(20)
```

```python
# Kutu grafii izimi
plt.figure()

# Seaborn ile boxplot (kutu grafii) iziliyor
# x = "features"  Her kutu bir zellii (rnein age, chol) temsil eder
# y = "value"  Bu zelliklerin leklenmi deerleri kullanlr
# hue = "output"  Her kutu, kalp hastal olan (1) ve olmayan (0) olarak renklendirilir
# data = data_melted  Veriyi uzun formata getirdiimiz DataFrame kullanlyor
sns.boxplot(x="features", y="value", hue="output", data=data_melted)

# Grafii ekranda gster
plt.show()
```

##  Aykr Deer Tespiti

* Aykr deer tespiti, veri biliminde, veri setindeki olaand veya beklenmedik gzlemleri belirlemek iin kullanlan bir yntemdir. Bu tespit, analizlerin doruluunu artrmak ve model performansn iyiletirmek asndan son derece nemlidir. Aykr deerler, genellikle verinin genel dalmn bozarak yanl sonular elde edilmesine neden olabilir. Dolaysyla, bu tr verilerin belirlenmesi ve gerektiinde karlmas, daha salam ve gvenilir analizler yaplmasna olanak tanr. Ayrca, aykr deerlerin incelenmesi, potansiyel frsatlar veya sorunlar hakknda nemli bilgiler sunarak, veri odakl karar verme srelerini destekler.

  

```python
# Saysal (numeric) deikenlerin isimleri bir listeye atanyor
# Bu deikenler, model eitimi veya istatistiksel analizler iin kullanlacak temel zelliklerdir
numeric_list = ["age", "trtbps", "chol", "thalachh", "oldpeak"]

# Orijinal veri setinden sadece bu saysal stunlar seerek yeni bir DataFrame oluturuluyor
# df.loc[:, numeric_list]  tm satrlar (:) ve sadece bu listedeki stunlar seilir
df_numeric = df.loc[:, numeric_list]

# Oluturulan saysal verilerden oluan DataFrame'in ilk 5 satr ekranda gsterilir
df_numeric.head()
```

```python
# df veri erevesindeki tm saysal stunlar iin temel istatistiksel zet bilgileri hesaplanr
df.describe()
```

```python
# Saysal stunlar arasnda dng balatlyor
for i in numeric_list:
    
    # Q1: 1. eyreklik (%25), Q3: 3. eyreklik (%75) hesaplanyor
    Q1 = np.percentile(df[i], 25)
    Q3 = np.percentile(df[i], 75)

    # IQR: Interquartile Range (eyrekler aras fark) hesaplanyor
    IQR = Q3 - Q1

    # Aykr deer eiklerinin st ve alt snrlar belirleniyor
    lower_bound = Q1 - 2.5 * IQR
    upper_bound = Q3 + 2.5 * IQR

    # Mevcut stunun ekli yazdrlyor (satr says)
    print(f"[{i}] Old shape: ", df[i].shape)

    # Aykr deerlerin indeksleri bulunuyor
    upper_outliers = df[df[i] > upper_bound].index
    lower_outliers = df[df[i] < lower_bound].index

    # Aykr deer indeksleri yazdrlyor
    print(f"[{i}] Upper outliers index: {list(upper_outliers)}")
    print(f"[{i}] Lower outliers index: {list(lower_outliers)}")

    # Aykr deerleri veri setinden kaldr
    df.drop(index=upper_outliers, inplace=True)
    df.drop(index=lower_outliers, inplace=True)

    # Yeni veri setinin ekli yazdrlyor
    print(f"[{i}] New shape: ", df.shape)
    print("-" * 50)
```

Modelleme, zellikle karmak ilikilerin ortaya konmasnda ve gelecekteki sonularn tahmin edilmesinde kritik bir rol oynar. rnein, bir iletmenin satlarnn gelecek dnemlerde nasl ekilleneceini tahmin etmek, bu sayede stratejik kararlar alabilmek iin nemlidir.

* Modelleme, veri bilimindeki en temel bileenlerden biridir nk doru bir model, verinin iindeki kalplar ve eilimleri ortaya kararak, daha bilinli kararlar alnmasna yardmc olur. Ayrca, modelleme sreci, veri setinin daha etkili bir ekilde kullanlmasn salar ve potansiyel sorunlar nceden belirlemeye yardmc olur.
```python
# Orijinal veri seti olan df'nin tm ierii kopyalanyor
# Bu sayede df1 zerinde yaplan ilemler, df'yi etkilemez (bamsz bir kopyadr)
df1 = df.copy(deep=True)
```

### Kategorik Stunlarn Kodlanmas

* Kategorik stunlarn kodlanmas, makine renimi ve veri analizi srelerinde, metin veya snflandrma gibi kategorik verilerin saysal forma dntrlmesidir. Bu ilem, algoritmalarn veriyi daha iyi anlamas ve ileyebilmesi iin gereklidir. rnein, "erkek" ve "kadn" gibi cinsiyet kategorileri, saysal deerlerle temsil edilerek modelin bu bilgiyi kullanmasn salar. Bu kodlama, modelin performansn artrarak daha doru tahminler yapmasn salar. Kategorik verilerin uygun ekilde kodlanmas, veri setinin kalitesini ykseltir ve analizin geerliliini artrr. Bu nedenle, veri hazrlama srecinin nemli bir parasdr.



```python
# df1 zerinde, belirlenen kategorik stunlar one-hot encoding yntemiyle saysal forma dntrlyor
# columns = categorical_list[:-1]  'output' dndaki tm kategorik stunlar dntrlr
# drop_first=True  lk kategori her stun iin drlr (referans olarak kabul edilir, multicollinearity engellenir)
df1 = pd.get_dummies(df1, columns=categorical_list[:-1], drop_first=True)

# Oluan yeni veri erevesinin ilk 5 satr grntleniyor
df1.head()
```

```python
# Bamsz deikenleri (features) ieren X veri erevesi oluturuluyor
# 'output' stunu karlarak kalan tm stunlar X'e atanr
X = df1.drop(columns=["output"], axis=1)

# Baml deiken (hedef deiken) olan 'output' ayr bir y deikenine atanr
# Bu stun DataFrame formatnda tutulur (istersen y = df1["output"] ile Series formatnda da alabilirsin)
y = df1[["output"]]
```

### lekleme

* lekleme, makine renimi ve veri analizi srelerinde, farkl leklerdeki zelliklerin belirli bir arala dntrlmesi ilemidir. Bu, zellikle mesafe tabanl algoritmalar iin kritik neme sahiptir; nk farkl leklerdeki deikenler, modelin renme srecini olumsuz etkileyebilir. rnein, bir zellik 0-1 aralnda, dieri ise 0-1000 aralnda olabilir. lekleme ile tm zellikler ayn lek altnda toplanarak, modelin daha iyi performans gstermesi salanr. Bylece, analiz ve tahmin sreleri daha gvenilir hale gelir. Bu nedenle, lekleme uygulamas, veri hazrlama aamasnda atlanmamas gereken nemli bir admdr.

```python
# StandardScaler snf, saysal verileri standardize etmek (leklemek) iin kullanlr
# Bu ilem, her bir zellii ortalamas 0 ve standart sapmas 1 olacak ekilde dntrr
# zellikle KNN, SVM, Lojistik Regresyon gibi lek duyarl algoritmalarda ok nemlidir
scaler = StandardScaler()

# scaler nesnesi oluturulduktan sonra yazdrldnda, varsaylan parametre ayarlar gsterilir
scaler
```

```python
# numeric_list[:-1]  Saysal stun listesinden 'output' hari tm deikenleri alr
# Bu stunlar standardize edilmek zere seilir (nk 'output' baml deikendir ve leklenmemelidir)

# Seilen saysal stunlar, StandardScaler ile dntrlr
# Bu ilem her bir stunu ortalamas 0, standart sapmas 1 olacak ekilde lekler
X[numeric_list[:-1]] = scaler.fit_transform(X[numeric_list[:-1]])

# leklendirilmi X veri setinin ilk 5 satrn gster
X.head()
```

### Eitim ve Test Ayrm

* Eitim ve test, makine renimi modellerinin performansn deerlendirmek iin veri setinin iki ayr ksma ayrlmas ilemidir. Bu yntem, verinin bir ksmnn modelin eitimi (train) iin, dier ksmnn ise modelin test edilmesi (test) iin kullanlmas anlamna gelir. Ama, modelin eitim srasnda rendiklerini yeni, grlmemi veriler zerinde nasl bir performans sergilediini lmektir. Bu sre, ar renmeyi (overfitting) nlemeye yardmc olur ve modelin genelleme yeteneini deerlendirir. Bu nedenle, doru bir train/test split, gvenilir sonular elde etmek iin kritik bir admdr ve verinin temsili asndan dikkatlice yaplmaldr.

```python
# Veri seti eitim ve test olarak ayrlyor
# test_size=0.1  verinin %10'u test, %90' eitim iin kullanlacak
# random_state=3  Ayn blmenin tekrar retilebilmesi iin sabit rastgelelik deeri veriliyor
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=3)

# Eitim ve test veri setlerinin boyutlar yazdrlyor
print("X_train shape :", X_train.shape)  # Eitim verisi (zellikler)
print("X_test shape  :", X_test.shape)   # Test verisi (zellikler)
print("y_train shape :", y_train.shape)  # Eitim verisi (etiket)
print("y_test shape  :", y_test.shape)   # Test verisi (etiket)
```

### Lojistik Regresyon

* Lojistik regresyon, iki veya daha fazla snf arasndaki olaslklar tahmin etmek iin kullanlan bir istatistiksel modeldir. zellikle, baml deikenin kategorik olduu durumlarda tercih edilir; rnein, bir olayn gerekleip gereklemeyecei gibi. Model, veriler arasndaki ilikileri renerek, yeni verilere dayanarak snflandrma yapar. Lojistik regresyon, basitlii ve yorumlanabilirlii sayesinde yaygn olarak kullanlr ve farkl alanlarda, zellikle salk, pazarlama ve finans gibi sektrlerde etkili sonular salar. Bu modelin kullanm, verilerin analizinde hzl ve etkili bir zm sunarak, karar verme srelerini destekler.

```python
# LogisticRegression snfndan bir model nesnesi oluturuluyor
# Bu model, ikili (binary) veya ok snfl snflandrma problemlerinde kullanlr
logreg = LogisticRegression()  # varsaylan hiperparametrelerle model balatlr

# Oluturulan model nesnesi yazdrlyor, bylece varsaylan parametreleri grebilirsin
logreg
```

```python
# Lojistik regresyon modeli, eitim verileri ile eitiliyor
# X_train: Girdi (bamsz) deikenler
# y_train: kt (baml) deiken  hedef snflar
logreg.fit(X_train, y_train)
```

```python
# Eitilmi model kullanlarak test verisinin her bir gzlemi iin snf olaslklar tahmin ediliyor
# predict_proba(), her gzlem iin her snfa ait olasl verir (rnein: [0.25, 0.75])
y_pred_prob = logreg.predict_proba(X_test)

# Olaslk tahminlerinin sonucu yazdrlr
y_pred_prob
```

```python
# Olaslklarn bulunduu y_pred_prob matrisinden, her gzlem iin en yksek olasla sahip snf seiliyor
# axis=1  her satr (gzlem) iin ilem yaplr
# np.argmax  en byk deerin (maksimum olasln) indeksini dner, bu da tahmin edilen snftr (0 veya 1)
y_pred = np.argmax(y_pred_prob, axis=1)

# Tahmin edilen snflar yazdrlr (rnein: [1 0 0 1 1 ...])
y_pred
```

```python
# Test veri kmesinde modelin doruluk (accuracy) skorunu hesaplyoruz

print("Test accuracy: {}".format(accuracy_score(y_pred, y_test)))
```

### ROC Erisi

* ROC (Receiver Operating Characteristic) erisi, bir snflandrma modelinin performansn deerlendirmek iin kullanlan bir grafiksel aratr. Bu eri, modelin doru pozitif orann (TPR) ve yanl pozitif orann (FPR) farkl eik deerleri iin gsterir. Ama, modelin eitli eik deerlerinde ne kadar iyi snflandrma yaptn grsel olarak deerlendirmektir. ROC erisinin altnda kalan alan (AUC), modelin genel baarsn len bir kriter olarak kullanlr; 1'e yakn deerler, mkemmel bir modeli ifade ederken, 0.5 deeri rastgele tahmin anlamna gelir. Bu nedenle, ROC erisi, model seimi ve deerlendirmesinde kritik bir rol oynar, veri bilimcilerin performans daha iyi anlamasn salar.

```python
# ROC erisi iin gerekli metrikler hesaplanyor:
# y_test: Gerek etiketler
# y_pred_prob[:, 1]: Pozitif snfa (snf=1) ait tahmin edilen olaslklar

# roc_curve(), farkl eik (threshold) deerlerine gre FPR, TPR ve eikleri hesaplar
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob[:, 1])
```

```python
# ROC erisinin referans izgisi (diyagonal): rastgele tahminin gstergesidir
# [0,1] - [0,1]  x ve y ekseninde 0'dan 1'e kadar uzanan dz bir izgi izer
# "k--"  siyah (k), kesikli (--) izgi
plt.plot([0, 1], [0, 1], 'k--', label='Rastgele Tahmin')

# Modelin ROC erisi iziliyor: X  FPR, Y  TPR
plt.plot(fpr, tpr, label="Lojistik Regresyon", color='blue', linewidth=2)

# X ve Y eksenine aklayc etiketler ekleniyor
plt.xlabel("False Positive Rate (Yanl Pozitif Oran)")
plt.ylabel("True Positive Rate (Doru Pozitif Oran)")

# Grafik bal belirleniyor
plt.title("ROC Erisi - Lojistik Regresyon")

# Aklama kutusu (legend) ekleniyor
plt.legend(loc="lower right")

# Grid izgileri eklenerek grafik okunabilir hale getiriliyor
plt.grid(True)

# Grafii ekranda gster
plt.show()
```

### Lojistik Regresyon Hiperparametre Ayar

* Lojistik regresyon hiperparametre ayar, modelin performansn optimize etmek amacyla yaplan bir sretir. Hiperparametreler, modelin eitim srecinde sabit kalan ve otomatik olarak renilmeyen parametrelerdir. rnein, dzenleme (regularization) tr ve katsaylar gibi ayarlar, modelin ar renme (overfitting) riskini azaltarak genelleme yeteneini artrabilir. Hiperparametre ayar, genellikle apraz dorulama (cross-validation) gibi yntemlerle gerekletirilir, bylece en iyi performans sunan parametre kombinasyonlar belirlenir. Bu sre, modelin doruluunu ve gvenilirliini artrarak, elde edilen sonularn daha salam olmasn salar.

```python
# LogisticRegression snfndan bir lojistik regresyon modeli nesnesi oluturuluyor
# Bu model, genellikle ikili snflandrma problemlerinde (rnein hastalk var/yok) kullanlr
# Varsaylan hiperparametrelerle balatlr
lr = LogisticRegression()

# Model nesnesi yazdrlarak kullanlan varsaylan parametreler grntlenir
lr
```

```python
# Kullanlacak ceza trleri (regularization penalties) bir liste halinde tanmlanyor
# "l1": L1 regularizasyonu (lasso)  baz katsaylar sfra indirerek zellik seimi salar
# "l2": L2 regularizasyonu (ridge)  byk katsaylar kltr, ancak sfrlamaz
penalty_options = ["l1", "l2"]

# GridSearch veya benzeri hiperparametre arama yntemleri iin parametre szl oluturuluyor
# Anahtar olarak 'penalty', deer olarak yukarda tanmlanan ceza trleri veriliyor
parameters = {"penalty": penalty_options}
```

```python
# GridSearchCV, verilen parametreler zerinden modelin en iyi versiyonunu bulmak iin apraz dorulama (cross-validation) uygular

# lr: Lojistik regresyon modeli (nceden tanmlanm)
# parameters: Denenecek ceza trlerini ieren parametre szl (rnein: {"penalty": ["l1", "l2"]})

# GridSearchCV nesnesi oluturuluyor
# cv=5 varsaylan olarak 5 katl apraz dorulama uygular (istee gre zelletirilebilir)
lr_searcher = GridSearchCV(estimator=lr, param_grid=parameters, cv=5)
```

```python
# GridSearchCV kullanlarak eitim verisi zerinden en iyi parametre kombinasyonu seiliyor
# X_train: Girdi (zellik) verileri
# y_train: Hedef (etiket) verileri
# Bu ilem, her parametre kombinasyonunu apraz dorulama (CV) ile deerlendirir

lr_searcher.fit(X_train, y_train)
```

```python
# GridSearchCV sonucunda en iyi bulunan hiperparametreler yazdrlyor
# best_params_  en yksek doruluu salayan parametre kombinasyonunu dndrr
print("Best Parameters Found by GridSearchCV:", lr_searcher.best_params_)
```

```python
# GridSearchCV sonucunda elde edilen en iyi model (best_estimator_) kullanlarak test verisi zerinde tahmin yaplyor
# X_test: Test veri kmesindeki zellikler
# y_pred: Test veri kmesindeki her bir gzlem iin tahmin edilen snflar (0 veya 1)
y_pred = lr_searcher.predict(X_test)
```

```python
# Test veri kmesinde modelin doruluk (accuracy) skoru hesaplanyor
# accuracy_score(): Tahmin edilen snflar (y_pred) ile gerek snflar (y_test) karlatrlr
test_accuracy = accuracy_score(y_test, y_pred)

# Hesaplanan doruluk deeri ekrana yzde formatnda yazdrlyor
print("Test Accuracy of the Best Model: {:.2f}%".format(test_accuracy * 100))
```
