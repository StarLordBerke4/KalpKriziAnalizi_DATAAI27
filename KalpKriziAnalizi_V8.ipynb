## ❤️ Kalp Krizi Hakkında

* Kalp krizi, **miyokard enfarktüsü** olarak da adlandırılır, kalp kasının bir kısmına yeterli kan gitmediğinde meydana gelir.
* Kan akışını geri kazandırmak için tedavi edilmeden geçen süre ne kadar uzun olursa, kalp kasına verilen hasar da o kadar büyük olur.
* **Koroner arter hastalığı (KAH)**, kalp krizinin başlıca nedenidir.
* KAH, kalbi besleyen atardamarların (koroner arterler) genellikle yağ ve kolesterol birikimiyle daralması veya tıkanması sonucu oluşur. Bu duruma **ateroskleroz** denir.
* Belirtiler genellikle göğüste ağrı veya rahatsızlık, nefes darlığı, kola (özellikle sol kola), çeneye, sırta veya mideye yayılan ağrı şeklinde kendini gösterir. Bazı durumlarda ise belirtiler daha hafif olabilir veya hiç görülmeyebilir (sessiz kalp krizi).
* Kalp krizi riskini artıran faktörler arasında **yüksek tansiyon, yüksek kolesterol, diyabet, obezite, sigara kullanımı, fiziksel hareketsizlik ve stres** yer alır. Genetik yatkınlık da önemli bir risk faktörüdür.
* Kalp krizi geçiren kişilerde hızlı tıbbi müdahale hayati önem taşır. Acil tedavi genellikle kan pıhtısını çözücü ilaçları (trombolitikler), balon anjiyoplasti veya stent yerleştirme gibi yöntemleri içerir.
* Kalp krizinden korunmak için **sağlıklı bir yaşam tarzı benimsemek, düzenli egzersiz yapmak, dengeli beslenmek ve risk faktörlerini yönetmek** büyük önem taşır.

## 🚀 Standardizasyon

* Veri bilimi standardizasyonu, verilerin tutarlılığını ve kalitesini artırmak amacıyla kullanılır. Bu süreç, farklı kaynaklardan gelen verilerin uyumlu hale getirilmesini sağlar, böylece analiz süreçleri daha verimli ve güvenilir olur.

```python
# StandardScaler sınıfı, verileri standart ölçeğe dönüştürmek için kullanılır
# Yani her bir özelliği (sütunu), ortalaması 0, standart sapması 1 olacak şekilde dönüştürür
scaler = StandardScaler()

# Oluşturulan scaler nesnesi çağrıldığında, içeriği (örneğin parametre ayarları) görüntülenir
scaler
```

```python
# StandardScaler nesnesi kullanılarak sayısal veriler standardize ediliyor (ölçekleniyor)
# df[numeric_list[:-1]] ifadesi: 'output' sütunu hariç, listedeki tüm sayısal sütunları seçer
# numeric_list[:-1] -> listenin son elemanı hariç (output hariç) olan kısmı
# scaler.fit_transform(...) -> veriyi hem öğrenir (fit), hem de dönüştürür (transform)
scaled_array = scaler.fit_transform(df[numeric_list[:-1]])
```

```python
# scaled_array değişkeni, StandardScaler ile ölçeklendirilmiş sayısal verileri içerir
# Bu veri, her sütunun ortalaması 0 ve standart sapması 1 olacak şekilde dönüştürülmüştür
# Bu satır çalıştırıldığında, scaled_array'in içeriği (NumPy dizisi olarak) ekrana yazdırılır
scaled_array
```

```python
# scaled_array verisi (NumPy dizisi) pandas DataFrame'e dönüştürülüyor
# Ardından describe() metodu ile temel istatistiksel özet bilgileri hesaplanıyor
# pd.DataFrame(scaled_array).describe()
```

## 📈 Kutu Grafiği Analizi

* Kutu grafiği, veri bilimi alanında verilerin dağılımını ve özet istatistiklerini görselleştirmek için kullanılan etkili bir araçtır. Özellikle, bir veri kümesinin merkezi eğilimi (ortalama veya medyan), yayılımı (çeyrekler arası aralık) ve olası aykırı değerleri hızlı bir şekilde anlamaya yardımcı olur. Kutu grafikleri, farklı gruplar arasındaki karşılaştırmaları kolaylaştırarak, verilerdeki farklılıkları ve benzerlikleri belirlemeye olanak tanır. Bu sayede, karar verme süreçlerinde daha bilinçli ve veri odaklı yaklaşımlar geliştirilir.

```python
# scaled_array verisi (NumPy dizisi), sütun adları belirtilerek pandas DataFrame'e dönüştürülüyor
# columns = numeric_list[:-1] -> 'output' hariç tüm sayısal sütun adları ekleniyor
df_dummy = pd.DataFrame(scaled_array, columns=numeric_list[:-1])

# Oluşturulan df_dummy veri çerçevesinin ilk 5 satırı ekranda gösteriliyor
df_dummy.head()
```

## 🔬 Aykırı Değer Tespiti

* Aykırı değer tespiti, veri biliminde, veri setindeki olağandışı veya beklenmedik gözlemleri belirlemek için kullanılan bir yöntemdir. Bu tespit, analizlerin doğruluğunu artırmak ve model performansını iyileştirmek açısından son derece önemlidir. Aykırı değerler, genellikle verinin genel dağılımını bozarak yanlış sonuçlar elde edilmesine neden olabilir. Dolayısıyla, bu tür verilerin belirlenmesi ve gerektiğinde çıkarılması, daha sağlam ve güvenilir analizler yapılmasına olanak tanır. Ayrıca, aykırı değerlerin incelenmesi, potansiyel fırsatlar veya sorunlar hakkında önemli bilgiler sunarak, veri odaklı karar verme süreçlerini destekler.

  

```python
# Sayısal (numeric) değişkenlerin isimleri bir listeye atanıyor
# Bu değişkenler, model eğitimi veya istatistiksel analizler için kullanılacak temel özelliklerdir
numeric_list = ["age", "trtbps", "chol", "thalachh", "oldpeak"]

# Orijinal veri setinden sadece bu sayısal sütunları seçerek yeni bir DataFrame oluşturuluyor
# df.loc[:, numeric_list] → tüm satırlar (:) ve sadece bu listedeki sütunlar seçilir
df_numeric = df.loc[:, numeric_list]

# Oluşturulan sayısal verilerden oluşan DataFrame'in ilk 5 satırı ekranda gösterilir
df_numeric.head()
```

```python
# df veri çerçevesindeki tüm sayısal sütunlar için temel istatistiksel özet bilgileri hesaplanır
df.describe()
```

```python
# Sayısal sütunlar arasında döngü başlatılıyor
for i in numeric_list:
    
    # Q1: 1. çeyreklik (%25), Q3: 3. çeyreklik (%75) hesaplanıyor
    Q1 = np.percentile(df[i], 25)
    Q3 = np.percentile(df[i], 75)

    # IQR: Interquartile Range (çeyrekler arası fark) hesaplanıyor
    IQR = Q3 - Q1

    # Aykırı değer eşiklerinin üst ve alt sınırları belirleniyor
    lower_bound = Q1 - 2.5 * IQR
    upper_bound = Q3 + 2.5 * IQR

    # Mevcut sütunun şekli yazdırılıyor (satır sayısı)
    print(f"[{i}] Old shape: ", df[i].shape)

    # Aykırı değerlerin indeksleri bulunuyor
    upper_outliers = df[df[i] > upper_bound].index
    lower_outliers = df[df[i] < lower_bound].index

    # Aykırı değer indeksleri yazdırılıyor
    print(f"[{i}] Upper outliers index: {list(upper_outliers)}")
    print(f"[{i}] Lower outliers index: {list(lower_outliers)}")

    # Aykırı değerleri veri setinden kaldır
    df.drop(index=upper_outliers, inplace=True)
    df.drop(index=lower_outliers, inplace=True)

    # Yeni veri setinin şekli yazdırılıyor
    print(f"[{i}] New shape: ", df.shape)
    print("-" * 50)
```

### Kategorik Sütunların Kodlanması

Bu işlem, algoritmaların veriyi daha iyi anlaması ve işleyebilmesi için gereklidir. Örneğin, "erkek" ve "kadın" gibi cinsiyet kategorileri, sayısal değerlerle temsil edilerek modelin bu bilgiyi kullanmasını sağlar. Bu kodlama, modelin performansını artırarak daha doğru tahminler yapmasını sağlar. Kategorik verilerin uygun şekilde kodlanması, veri setinin kalitesini yükseltir ve analizin geçerliliğini artırır. Bu nedenle, veri hazırlama sürecinin önemli bir parçasıdır.



```python
# df1 üzerinde, belirlenen kategorik sütunlar one-hot encoding yöntemiyle sayısal forma dönüştürülüyor
# columns = categorical_list[:-1] → 'output' dışındaki tüm kategorik sütunlar dönüştürülür
# drop_first=True → İlk kategori her sütun için düşürülür (referans olarak kabul edilir, multicollinearity engellenir)
df1 = pd.get_dummies(df1, columns=categorical_list[:-1], drop_first=True)

# Oluşan yeni veri çerçevesinin ilk 5 satırı görüntüleniyor
df1.head()
```

```python
# Bağımsız değişkenleri (features) içeren X veri çerçevesi oluşturuluyor
# 'output' sütunu çıkarılarak kalan tüm sütunlar X'e atanır
X = df1.drop(columns=["output"], axis=1)

# Bağımlı değişken (hedef değişken) olan 'output' ayrı bir y değişkenine atanır
# Bu sütun DataFrame formatında tutulur (istersen y = df1["output"] ile Series formatında da alabilirsin)
y = df1[["output"]]
```

### Ölçekleme

* Ölçekleme, makine öğrenimi ve veri analizi süreçlerinde, farklı ölçeklerdeki özelliklerin belirli bir aralığa dönüştürülmesi işlemidir. Bu, özellikle mesafe tabanlı algoritmalar için kritik öneme sahiptir; çünkü farklı ölçeklerdeki değişkenler, modelin öğrenme sürecini olumsuz etkileyebilir. Örneğin, bir özellik 0-1 aralığında, diğeri ise 0-1000 aralığında olabilir. Ölçekleme ile tüm özellikler aynı ölçek altında toplanarak, modelin daha iyi performans göstermesi sağlanır. Böylece, analiz ve tahmin süreçleri daha güvenilir hale gelir. Bu nedenle, ölçekleme uygulaması, veri hazırlama aşamasında atlanmaması gereken önemli bir adımdır.

```python
# StandardScaler sınıfı, sayısal verileri standardize etmek (ölçeklemek) için kullanılır
# Bu işlem, her bir özelliği ortalaması 0 ve standart sapması 1 olacak şekilde dönüştürür
# Özellikle KNN, SVM, Lojistik Regresyon gibi ölçek duyarlı algoritmalarda çok önemlidir
scaler = StandardScaler()

# scaler nesnesi oluşturulduktan sonra yazdırıldığında, varsayılan parametre ayarları gösterilir
scaler
```

```python
# numeric_list[:-1] → Sayısal sütun listesinden 'output' hariç tüm değişkenleri alır
# Bu sütunlar standardize edilmek üzere seçilir (çünkü 'output' bağımlı değişkendir ve ölçeklenmemelidir)

# Seçilen sayısal sütunlar, StandardScaler ile dönüştürülür
# Bu işlem her bir sütunu ortalaması 0, standart sapması 1 olacak şekilde ölçekler
X[numeric_list[:-1]] = scaler.fit_transform(X[numeric_list[:-1]])

# Ölçeklendirilmiş X veri setinin ilk 5 satırını göster
X.head()
```

### Eğitim ve Test Ayrımı

* Eğitim ve test, makine öğrenimi modellerinin performansını değerlendirmek için veri setinin iki ayrı kısma ayrılması işlemidir. Bu yöntem, verinin bir kısmının modelin eğitimi (train) için, diğer kısmının ise modelin test edilmesi (test) için kullanılması anlamına gelir. Amaç, modelin eğitim sırasında öğrendiklerini yeni, görülmemiş veriler üzerinde nasıl bir performans sergilediğini ölçmektir. Bu süreç, aşırı öğrenmeyi (overfitting) önlemeye yardımcı olur ve modelin genelleme yeteneğini değerlendirir. Bu nedenle, doğru bir train/test split, güvenilir sonuçlar elde etmek için kritik bir adımdır ve verinin temsili açısından dikkatlice yapılmalıdır.

```python
# Veri seti eğitim ve test olarak ayrılıyor
# test_size=0.1 → verinin %10'u test, %90'ı eğitim için kullanılacak
# random_state=3 → Aynı bölmenin tekrar üretilebilmesi için sabit rastgelelik değeri veriliyor
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=3)

# Eğitim ve test veri setlerinin boyutları yazdırılıyor
print("X_train shape :", X_train.shape)  # Eğitim verisi (özellikler)
print("X_test shape  :", X_test.shape)   # Test verisi (özellikler)
print("y_train shape :", y_train.shape)  # Eğitim verisi (etiket)
print("y_test shape  :", y_test.shape)   # Test verisi (etiket)
```

### Lojistik Regresyon

* Lojistik regresyon, iki veya daha fazla sınıf arasındaki olasılıkları tahmin etmek için kullanılan bir istatistiksel modeldir. Özellikle, bağımlı değişkenin kategorik olduğu durumlarda tercih edilir; örneğin, bir olayın gerçekleşip gerçekleşmeyeceği gibi. Model, veriler arasındaki ilişkileri öğrenerek, yeni verilere dayanarak sınıflandırma yapar. Lojistik regresyon, basitliği ve yorumlanabilirliği sayesinde yaygın olarak kullanılır ve farklı alanlarda, özellikle sağlık, pazarlama ve finans gibi sektörlerde etkili sonuçlar sağlar. Bu modelin kullanımı, verilerin analizinde hızlı ve etkili bir çözüm sunarak, karar verme süreçlerini destekler.

```python
# LogisticRegression sınıfından bir model nesnesi oluşturuluyor
# Bu model, ikili (binary) veya çok sınıflı sınıflandırma problemlerinde kullanılır
logreg = LogisticRegression()  # varsayılan hiperparametrelerle model başlatılır

# Oluşturulan model nesnesi yazdırılıyor, böylece varsayılan parametreleri görebilirsin
logreg
```

```python
# Lojistik regresyon modeli, eğitim verileri ile eğitiliyor
# X_train: Girdi (bağımsız) değişkenler
# y_train: Çıktı (bağımlı) değişken — hedef sınıflar
logreg.fit(X_train, y_train)
```

```python
# Eğitilmiş model kullanılarak test verisinin her bir gözlemi için sınıf olasılıkları tahmin ediliyor
# predict_proba(), her gözlem için her sınıfa ait olasılığı verir (örneğin: [0.25, 0.75])
y_pred_prob = logreg.predict_proba(X_test)

# Olasılık tahminlerinin sonucu yazdırılır
y_pred_prob
```

```python
# Olasılıkların bulunduğu y_pred_prob matrisinden, her gözlem için en yüksek olasılığa sahip sınıf seçiliyor
# axis=1 → her satır (gözlem) için işlem yapılır
# np.argmax → en büyük değerin (maksimum olasılığın) indeksini döner, bu da tahmin edilen sınıftır (0 veya 1)
y_pred = np.argmax(y_pred_prob, axis=1)

# Tahmin edilen sınıflar yazdırılır (örneğin: [1 0 0 1 1 ...])
y_pred
```

```python
# Test veri kümesinde modelin doğruluk (accuracy) skorunu hesaplıyoruz

print("Test accuracy: {}".format(accuracy_score(y_pred, y_test)))
```

### ROC Eğrisi

* ROC (Receiver Operating Characteristic) eğrisi, bir sınıflandırma modelinin performansını değerlendirmek için kullanılan bir grafiksel araçtır. Bu eğri, modelin doğru pozitif oranını (TPR) ve yanlış pozitif oranını (FPR) farklı eşik değerleri için gösterir. Amaç, modelin çeşitli eşik değerlerinde ne kadar iyi sınıflandırma yaptığını görsel olarak değerlendirmektir. ROC eğrisinin altında kalan alan (AUC), modelin genel başarısını ölçen bir kriter olarak kullanılır; 1'e yakın değerler, mükemmel bir modeli ifade ederken, 0.5 değeri rastgele tahmin anlamına gelir. Bu nedenle, ROC eğrisi, model seçimi ve değerlendirmesinde kritik bir rol oynar, veri bilimcilerin performansı daha iyi anlamasını sağlar.

```python
# ROC eğrisi için gerekli metrikler hesaplanıyor:
# y_test: Gerçek etiketler
# y_pred_prob[:, 1]: Pozitif sınıfa (sınıf=1) ait tahmin edilen olasılıklar

# roc_curve(), farklı eşik (threshold) değerlerine göre FPR, TPR ve eşikleri hesaplar
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob[:, 1])
```

```python
# ROC eğrisinin referans çizgisi (diyagonal): rastgele tahminin göstergesidir
# [0,1] - [0,1] → x ve y ekseninde 0'dan 1'e kadar uzanan düz bir çizgi çizer
# "k--" → siyah (k), kesikli (--) çizgi
plt.plot([0, 1], [0, 1], 'k--', label='Rastgele Tahmin')

# Modelin ROC eğrisi çiziliyor: X → FPR, Y → TPR
plt.plot(fpr, tpr, label="Lojistik Regresyon", color='blue', linewidth=2)

# X ve Y eksenine açıklayıcı etiketler ekleniyor
plt.xlabel("False Positive Rate (Yanlış Pozitif Oranı)")
plt.ylabel("True Positive Rate (Doğru Pozitif Oranı)")

# Grafik başlığı belirleniyor
plt.title("ROC Eğrisi - Lojistik Regresyon")

# Açıklama kutusu (legend) ekleniyor
plt.legend(loc="lower right")

# Grid çizgileri eklenerek grafik okunabilir hale getiriliyor
plt.grid(True)

# Grafiği ekranda göster
plt.show()
```

### Lojistik Regresyon Hiperparametre Ayarı

* Lojistik regresyon hiperparametre ayarı, modelin performansını optimize etmek amacıyla yapılan bir süreçtir. Hiperparametreler, modelin eğitim sürecinde sabit kalan ve otomatik olarak öğrenilmeyen parametrelerdir. Örneğin, düzenleme (regularization) türü ve katsayıları gibi ayarlar, modelin aşırı öğrenme (overfitting) riskini azaltarak genelleme yeteneğini artırabilir. Hiperparametre ayarı, genellikle çapraz doğrulama (cross-validation) gibi yöntemlerle gerçekleştirilir, böylece en iyi performansı sunan parametre kombinasyonları belirlenir. Bu süreç, modelin doğruluğunu ve güvenilirliğini artırarak, elde edilen sonuçların daha sağlam olmasını sağlar.

```python
# LogisticRegression sınıfından bir lojistik regresyon modeli nesnesi oluşturuluyor
# Bu model, genellikle ikili sınıflandırma problemlerinde (örneğin hastalık var/yok) kullanılır
# Varsayılan hiperparametrelerle başlatılır
lr = LogisticRegression()

# Model nesnesi yazdırılarak kullanılan varsayılan parametreler görüntülenir
lr
```

```python
# Kullanılacak ceza türleri (regularization penalties) bir liste halinde tanımlanıyor
# "l1": L1 regularizasyonu (lasso) → bazı katsayıları sıfıra indirerek özellik seçimi sağlar
# "l2": L2 regularizasyonu (ridge) → büyük katsayıları küçültür, ancak sıfırlamaz
penalty_options = ["l1", "l2"]

# GridSearch veya benzeri hiperparametre arama yöntemleri için parametre sözlüğü oluşturuluyor
# Anahtar olarak 'penalty', değer olarak yukarıda tanımlanan ceza türleri veriliyor
parameters = {"penalty": penalty_options}
```

```python
# GridSearchCV, verilen parametreler üzerinden modelin en iyi versiyonunu bulmak için çapraz doğrulama (cross-validation) uygular

# lr: Lojistik regresyon modeli (önceden tanımlanmış)
# parameters: Denenecek ceza türlerini içeren parametre sözlüğü (örneğin: {"penalty": ["l1", "l2"]})

# GridSearchCV nesnesi oluşturuluyor
# cv=5 varsayılan olarak 5 katlı çapraz doğrulama uygular (isteğe göre özelleştirilebilir)
lr_searcher = GridSearchCV(estimator=lr, param_grid=parameters, cv=5)
```

```python
# GridSearchCV kullanılarak eğitim verisi üzerinden en iyi parametre kombinasyonu seçiliyor
# X_train: Girdi (özellik) verileri
# y_train: Hedef (etiket) verileri
# Bu işlem, her parametre kombinasyonunu çapraz doğrulama (CV) ile değerlendirir

lr_searcher.fit(X_train, y_train)
```

```python
# GridSearchCV sonucunda en iyi bulunan hiperparametreler yazdırılıyor
# best_params_ → en yüksek doğruluğu sağlayan parametre kombinasyonunu döndürür
print("Best Parameters Found by GridSearchCV:", lr_searcher.best_params_)
```

```python
# GridSearchCV sonucunda elde edilen en iyi model (best_estimator_) kullanılarak test verisi üzerinde tahmin yapılıyor
# X_test: Test veri kümesindeki özellikler
# y_pred: Test veri kümesindeki her bir gözlem için tahmin edilen sınıflar (0 veya 1)
y_pred = lr_searcher.predict(X_test)
```

```python
# Test veri kümesinde modelin doğruluk (accuracy) skoru hesaplanıyor
# accuracy_score(): Tahmin edilen sınıflar (y_pred) ile gerçek sınıflar (y_test) karşılaştırılır
test_accuracy = accuracy_score(y_test, y_pred)

# Hesaplanan doğruluk değeri ekrana yüzde formatında yazdırılıyor
print("Test Accuracy of the Best Model: {:.2f}%".format(test_accuracy * 100))
```
