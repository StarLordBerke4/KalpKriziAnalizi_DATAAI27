## ðŸ”¬ AykÄ±rÄ± DeÄŸer Tespiti

* AykÄ±rÄ± deÄŸer tespiti, veri biliminde, veri setindeki olaÄŸandÄ±ÅŸÄ± veya beklenmedik gÃ¶zlemleri belirlemek iÃ§in kullanÄ±lan bir yÃ¶ntemdir. Bu tespit, analizlerin doÄŸruluÄŸunu artÄ±rmak ve model performansÄ±nÄ± iyileÅŸtirmek aÃ§Ä±sÄ±ndan son derece Ã¶nemlidir. AykÄ±rÄ± deÄŸerler, genellikle verinin genel daÄŸÄ±lÄ±mÄ±nÄ± bozarak yanlÄ±ÅŸ sonuÃ§lar elde edilmesine neden olabilir. DolayÄ±sÄ±yla, bu tÃ¼r verilerin belirlenmesi ve gerektiÄŸinde Ã§Ä±karÄ±lmasÄ±, daha saÄŸlam ve gÃ¼venilir analizler yapÄ±lmasÄ±na olanak tanÄ±r. AyrÄ±ca, aykÄ±rÄ± deÄŸerlerin incelenmesi, potansiyel fÄ±rsatlar veya sorunlar hakkÄ±nda Ã¶nemli bilgiler sunarak, veri odaklÄ± karar verme sÃ¼reÃ§lerini destekler.

  

```python
# SayÄ±sal (numeric) deÄŸiÅŸkenlerin isimleri bir listeye atanÄ±yor
# Bu deÄŸiÅŸkenler, model eÄŸitimi veya istatistiksel analizler iÃ§in kullanÄ±lacak temel Ã¶zelliklerdir
numeric_list = ["age", "trtbps", "chol", "thalachh", "oldpeak"]

# Orijinal veri setinden sadece bu sayÄ±sal sÃ¼tunlarÄ± seÃ§erek yeni bir DataFrame oluÅŸturuluyor
# df.loc[:, numeric_list] â†’ tÃ¼m satÄ±rlar (:) ve sadece bu listedeki sÃ¼tunlar seÃ§ilir
df_numeric = df.loc[:, numeric_list]

# OluÅŸturulan sayÄ±sal verilerden oluÅŸan DataFrame'in ilk 5 satÄ±rÄ± ekranda gÃ¶sterilir
df_numeric.head()
```

```python
# df veri Ã§erÃ§evesindeki tÃ¼m sayÄ±sal sÃ¼tunlar iÃ§in temel istatistiksel Ã¶zet bilgileri hesaplanÄ±r
df.describe()
```

```python
# SayÄ±sal sÃ¼tunlar arasÄ±nda dÃ¶ngÃ¼ baÅŸlatÄ±lÄ±yor
for i in numeric_list:
    
    # Q1: 1. Ã§eyreklik (%25), Q3: 3. Ã§eyreklik (%75) hesaplanÄ±yor
    Q1 = np.percentile(df[i], 25)
    Q3 = np.percentile(df[i], 75)

    # IQR: Interquartile Range (Ã§eyrekler arasÄ± fark) hesaplanÄ±yor
    IQR = Q3 - Q1

    # AykÄ±rÄ± deÄŸer eÅŸiklerinin Ã¼st ve alt sÄ±nÄ±rlarÄ± belirleniyor
    lower_bound = Q1 - 2.5 * IQR
    upper_bound = Q3 + 2.5 * IQR

    # Mevcut sÃ¼tunun ÅŸekli yazdÄ±rÄ±lÄ±yor (satÄ±r sayÄ±sÄ±)
    print(f"[{i}] Old shape: ", df[i].shape)

    # AykÄ±rÄ± deÄŸerlerin indeksleri bulunuyor
    upper_outliers = df[df[i] > upper_bound].index
    lower_outliers = df[df[i] < lower_bound].index

    # AykÄ±rÄ± deÄŸer indeksleri yazdÄ±rÄ±lÄ±yor
    print(f"[{i}] Upper outliers index: {list(upper_outliers)}")
    print(f"[{i}] Lower outliers index: {list(lower_outliers)}")

    # AykÄ±rÄ± deÄŸerleri veri setinden kaldÄ±r
    df.drop(index=upper_outliers, inplace=True)
    df.drop(index=lower_outliers, inplace=True)

    # Yeni veri setinin ÅŸekli yazdÄ±rÄ±lÄ±yor
    print(f"[{i}] New shape: ", df.shape)
    print("-" * 50)
```

### Lojistik Regresyon

* Lojistik regresyon, iki veya daha fazla sÄ±nÄ±f arasÄ±ndaki olasÄ±lÄ±klarÄ± tahmin etmek iÃ§in kullanÄ±lan bir istatistiksel modeldir. Ã–zellikle, baÄŸÄ±mlÄ± deÄŸiÅŸkenin kategorik olduÄŸu durumlarda tercih edilir; Ã¶rneÄŸin, bir olayÄ±n gerÃ§ekleÅŸip gerÃ§ekleÅŸmeyeceÄŸi gibi. Model, veriler arasÄ±ndaki iliÅŸkileri Ã¶ÄŸrenerek, yeni verilere dayanarak sÄ±nÄ±flandÄ±rma yapar.
```python
# LogisticRegression sÄ±nÄ±fÄ±ndan bir model nesnesi oluÅŸturuluyor
# Bu model, ikili (binary) veya Ã§ok sÄ±nÄ±flÄ± sÄ±nÄ±flandÄ±rma problemlerinde kullanÄ±lÄ±r
logreg = LogisticRegression()  # varsayÄ±lan hiperparametrelerle model baÅŸlatÄ±lÄ±r

# OluÅŸturulan model nesnesi yazdÄ±rÄ±lÄ±yor, bÃ¶ylece varsayÄ±lan parametreleri gÃ¶rebilirsin
logreg
```

### ROC EÄŸrisi

* ROC (Receiver Operating Characteristic) eÄŸrisi, bir sÄ±nÄ±flandÄ±rma modelinin performansÄ±nÄ± deÄŸerlendirmek iÃ§in kullanÄ±lan bir grafiksel araÃ§tÄ±r. Bu eÄŸri, modelin doÄŸru pozitif oranÄ±nÄ± (TPR) ve yanlÄ±ÅŸ pozitif oranÄ±nÄ± (FPR) farklÄ± eÅŸik deÄŸerleri iÃ§in gÃ¶sterir. AmaÃ§, modelin Ã§eÅŸitli eÅŸik deÄŸerlerinde ne kadar iyi sÄ±nÄ±flandÄ±rma yaptÄ±ÄŸÄ±nÄ± gÃ¶rsel olarak deÄŸerlendirmektir. ROC eÄŸrisinin altÄ±nda kalan alan (AUC), modelin genel baÅŸarÄ±sÄ±nÄ± Ã¶lÃ§en bir kriter olarak kullanÄ±lÄ±r; 1'e yakÄ±n deÄŸerler, mÃ¼kemmel bir modeli ifade ederken, 0.5 deÄŸeri rastgele tahmin anlamÄ±na gelir. Bu nedenle, ROC eÄŸrisi, model seÃ§imi ve deÄŸerlendirmesinde kritik bir rol oynar, veri bilimcilerin performansÄ± daha iyi anlamasÄ±nÄ± saÄŸlar.

```python
# ROC eÄŸrisi iÃ§in gerekli metrikler hesaplanÄ±yor:
# y_test: GerÃ§ek etiketler
# y_pred_prob[:, 1]: Pozitif sÄ±nÄ±fa (sÄ±nÄ±f=1) ait tahmin edilen olasÄ±lÄ±klar

# roc_curve(), farklÄ± eÅŸik (threshold) deÄŸerlerine gÃ¶re FPR, TPR ve eÅŸikleri hesaplar
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob[:, 1])
```

```python
# ROC eÄŸrisinin referans Ã§izgisi (diyagonal): rastgele tahminin gÃ¶stergesidir
# [0,1] - [0,1] â†’ x ve y ekseninde 0'dan 1'e kadar uzanan dÃ¼z bir Ã§izgi Ã§izer
# "k--" â†’ siyah (k), kesikli (--) Ã§izgi
plt.plot([0, 1], [0, 1], 'k--', label='Rastgele Tahmin')

# Modelin ROC eÄŸrisi Ã§iziliyor: X â†’ FPR, Y â†’ TPR
plt.plot(fpr, tpr, label="Lojistik Regresyon", color='blue', linewidth=2)

# X ve Y eksenine aÃ§Ä±klayÄ±cÄ± etiketler ekleniyor
plt.xlabel("False Positive Rate (YanlÄ±ÅŸ Pozitif OranÄ±)")
plt.ylabel("True Positive Rate (DoÄŸru Pozitif OranÄ±)")

# Grafik baÅŸlÄ±ÄŸÄ± belirleniyor
plt.title("ROC EÄŸrisi - Lojistik Regresyon")

# AÃ§Ä±klama kutusu (legend) ekleniyor
plt.legend(loc="lower right")

# Grid Ã§izgileri eklenerek grafik okunabilir hale getiriliyor
plt.grid(True)

# GrafiÄŸi ekranda gÃ¶ster
plt.show()
```

### Lojistik Regresyon Hiperparametre AyarÄ±

* Lojistik regresyon hiperparametre ayarÄ±, modelin performansÄ±nÄ± optimize etmek amacÄ±yla yapÄ±lan bir sÃ¼reÃ§tir. Hiperparametreler, modelin eÄŸitim sÃ¼recinde sabit kalan ve otomatik olarak Ã¶ÄŸrenilmeyen parametrelerdir. Ã–rneÄŸin, dÃ¼zenleme (regularization) tÃ¼rÃ¼ ve katsayÄ±larÄ± gibi ayarlar, modelin aÅŸÄ±rÄ± Ã¶ÄŸrenme (overfitting) riskini azaltarak genelleme yeteneÄŸini artÄ±rabilir. Hiperparametre ayarÄ±, genellikle Ã§apraz doÄŸrulama (cross-validation) gibi yÃ¶ntemlerle gerÃ§ekleÅŸtirilir, bÃ¶ylece en iyi performansÄ± sunan parametre kombinasyonlarÄ± belirlenir. Bu sÃ¼reÃ§, modelin doÄŸruluÄŸunu ve gÃ¼venilirliÄŸini artÄ±rarak, elde edilen sonuÃ§larÄ±n daha saÄŸlam olmasÄ±nÄ± saÄŸlar.

```python
# LogisticRegression sÄ±nÄ±fÄ±ndan bir lojistik regresyon modeli nesnesi oluÅŸturuluyor
# Bu model, genellikle ikili sÄ±nÄ±flandÄ±rma problemlerinde (Ã¶rneÄŸin hastalÄ±k var/yok) kullanÄ±lÄ±r
# VarsayÄ±lan hiperparametrelerle baÅŸlatÄ±lÄ±r
lr = LogisticRegression()

# Model nesnesi yazdÄ±rÄ±larak kullanÄ±lan varsayÄ±lan parametreler gÃ¶rÃ¼ntÃ¼lenir
lr
```

```python
# KullanÄ±lacak ceza tÃ¼rleri (regularization penalties) bir liste halinde tanÄ±mlanÄ±yor
# "l1": L1 regularizasyonu (lasso) â†’ bazÄ± katsayÄ±larÄ± sÄ±fÄ±ra indirerek Ã¶zellik seÃ§imi saÄŸlar
# "l2": L2 regularizasyonu (ridge) â†’ bÃ¼yÃ¼k katsayÄ±larÄ± kÃ¼Ã§Ã¼ltÃ¼r, ancak sÄ±fÄ±rlamaz
penalty_options = ["l1", "l2"]

# GridSearch veya benzeri hiperparametre arama yÃ¶ntemleri iÃ§in parametre sÃ¶zlÃ¼ÄŸÃ¼ oluÅŸturuluyor
# Anahtar olarak 'penalty', deÄŸer olarak yukarÄ±da tanÄ±mlanan ceza tÃ¼rleri veriliyor
parameters = {"penalty": penalty_options}
```

```python
# GridSearchCV, verilen parametreler Ã¼zerinden modelin en iyi versiyonunu bulmak iÃ§in Ã§apraz doÄŸrulama (cross-validation) uygular

# lr: Lojistik regresyon modeli (Ã¶nceden tanÄ±mlanmÄ±ÅŸ)
# parameters: Denenecek ceza tÃ¼rlerini iÃ§eren parametre sÃ¶zlÃ¼ÄŸÃ¼ (Ã¶rneÄŸin: {"penalty": ["l1", "l2"]})

# GridSearchCV nesnesi oluÅŸturuluyor
# cv=5 varsayÄ±lan olarak 5 katlÄ± Ã§apraz doÄŸrulama uygular (isteÄŸe gÃ¶re Ã¶zelleÅŸtirilebilir)
lr_searcher = GridSearchCV(estimator=lr, param_grid=parameters, cv=5)
```

```python
# GridSearchCV kullanÄ±larak eÄŸitim verisi Ã¼zerinden en iyi parametre kombinasyonu seÃ§iliyor
# X_train: Girdi (Ã¶zellik) verileri
# y_train: Hedef (etiket) verileri
# Bu iÅŸlem, her parametre kombinasyonunu Ã§apraz doÄŸrulama (CV) ile deÄŸerlendirir

lr_searcher.fit(X_train, y_train)
```

```python
# GridSearchCV sonucunda en iyi bulunan hiperparametreler yazdÄ±rÄ±lÄ±yor
# best_params_ â†’ en yÃ¼ksek doÄŸruluÄŸu saÄŸlayan parametre kombinasyonunu dÃ¶ndÃ¼rÃ¼r
print("Best Parameters Found by GridSearchCV:", lr_searcher.best_params_)
```

```python
# GridSearchCV sonucunda elde edilen en iyi model (best_estimator_) kullanÄ±larak test verisi Ã¼zerinde tahmin yapÄ±lÄ±yor
# X_test: Test veri kÃ¼mesindeki Ã¶zellikler
# y_pred: Test veri kÃ¼mesindeki her bir gÃ¶zlem iÃ§in tahmin edilen sÄ±nÄ±flar (0 veya 1)
y_pred = lr_searcher.predict(X_test)
```

```python
# Test veri kÃ¼mesinde modelin doÄŸruluk (accuracy) skoru hesaplanÄ±yor
# accuracy_score(): Tahmin edilen sÄ±nÄ±flar (y_pred) ile gerÃ§ek sÄ±nÄ±flar (y_test) karÅŸÄ±laÅŸtÄ±rÄ±lÄ±r
test_accuracy = accuracy_score(y_test, y_pred)

# Hesaplanan doÄŸruluk deÄŸeri ekrana yÃ¼zde formatÄ±nda yazdÄ±rÄ±lÄ±yor
print("Test Accuracy of the Best Model: {:.2f}%".format(test_accuracy * 100))
```
